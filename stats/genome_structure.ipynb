```python
""" run this cell to import structure data """

# import genome structure data
import marshal
marshal_filename = "../master_table/genome_structure.data.marshal"
gsdata = open(marshal_filename,'rb')
genome_structures = marshal.load(gsdata)
dgs = dict(genome_structures)
gsdata.close()

```

```python
""" the real code lives here (run the cell above once too to import data) """

#import what to plot
# format: a text file with
#  [accession] [otu_name]
to_plot = set()
for line in open("../master_table/genome_structure.to_plot.txt"):
    to_plot.add(tuple(line.split()))
#print(to_plot)

from dna_features_viewer import GraphicFeature, GraphicRecord
#palette = sns.color_palette("hls", len(data))
palette = sns.color_palette("cubehelix", len(data))
cur_color = 0
dpalette = dict()

import matplotlib.pyplot as plt
nb_plots = len(to_plot)
fig, subplots  = plt.subplots(
    nb_plots, 1, figsize=(20, 3*nb_plots), sharex=True)
subplot_i = 0
for accession, why in sorted(list(to_plot),key = lambda x:x[1])[:nb_plots]:
    features = []
    for i,feature in enumerate(dgs[accession]):
        start, end, name, ctg, is_complete = feature
        if name not in dpalette:
            dpalette[name] = palette[cur_color] # use next available color in palette
            cur_color = (cur_color + 1) % len(data)
        color = dpalette[name]
        features += [GraphicFeature(start=start, end=end, strand=+1, 
                                    color = color,#)]
                                    label=name)]
                    # color="#ffd700")]

    record = GraphicRecord(sequence_length=30000, feature_level_height=1, features=features)
    ax, _ = record.plot(ax=subplots[subplot_i], figure_width=30)
    ax.set_facecolor("white")
    ax.set_title('accession ' + accession + ' OTU: ' + why,loc='left')
    subplot_i += 1
fig.tight_layout(pad=5.)
fig.savefig('genome_structure.pdf', bbox_inches='tight')
```
